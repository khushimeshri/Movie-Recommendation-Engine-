```python
# Fill missing values in selected features with empty strings
for feature in features_selected:
  movies_data[feature] = movies_data[feature].fillna('')
display (movies_data.head())
```

```python
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
display (combined_features)
```

```python
#Creating feature vectors using TF-IDF vectorizer
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
display (feature_vectors.shape)
print (feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
print  (similarity )
```

```python
 #Get user input 
movie_name = input('Enter name of your favorite movie : ')
print(movie_name)
```

```python
All_titles=movies_data['title'].tolist()
print(All_titles)
```

```python
#Finding the closest matches to the user's input
find_close_match = difflib.get_close_matches(movie_name, All_titles)
print(find_close_match)
close_match = find_close_match[0]
print(close_match)
```

```python
index= movies_data[movies_data.title == close_match]['index'].values[0]
print(index)
```

```python
similarity_score = list(enumerate(similarity[index]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)#The closest match will be at the top and is sorted first i.e. at index 26
```

```python
print('Movies suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<10):
    print(i, '-->',title_from_index)
    i+=1
```

```python
#or using iloc function 
def recommend(movies):
    index=movies_data[movies_data['title']==close_match].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:10]:
        print(movies_data.iloc[i[0]].title)
```
